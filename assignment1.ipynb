The processes mainly included to reading datasets, removing unnecessary and repeated features and merging datasets.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## World Population dataset

```python
world=pd.read_csv('/kaggle/input/world-population-dataset/world_population.csv')
world.columns
```

```python
world.drop(['Rank', "CCA3", '2010 Population', '2000 Population',
       '1990 Population', '1980 Population', '1970 Population'], axis=1, inplace=True)
world.rename(columns={'Country/Territory': 'Country', 'Density (per km²)':'Destiny',
                      'Area (km²)':"Area"}, inplace=True)
world
```

```python
world.isna().sum()
```

## World happiness dataset

```python
happiness=pd.read_csv('/kaggle/input/world-happiness/2019.csv')
happiness.columns
```

```python
happiness.rename(columns={'Country or region': 'Country', 'Score':'happiness'},
                 inplace=True)
happiness.drop(['Overall rank'], axis=1, inplace=True)
happiness
```

```python
happiness.sort_values(by=['Country'], inplace=True)
happiness = happiness.reset_index()
happiness.drop('index', axis=1, inplace=True)
```

```python
happiness.isna().sum()
```

## Countries of the world dataset

```python
countries=pd.read_csv('/kaggle/input/countries-of-the-world/countries of the world.csv')
countries.columns
```

```python
#countries.rename(columns={'Country': 'country'}, inplace=True)
countries.drop(['Region', 'Population', 'Area (sq. mi.)','Pop. Density (per sq. mi.)', 'GDP ($ per capita)'],
             axis=1, inplace=True)
countries
```

```python
countries.isna().sum()
```

When I am merging other datasets with this dataset, I am having mindblowing errors. With little bit of digging, I found that for each of country name, they included white-space at the end. So, I use python's right strip.

```python
new_list=[]
for country in countries.Country.tolist():
    new_list.append(country.rstrip())
countries.Country=new_list
```

## 3. Best predictor of smartness

For this time, I wanted to know about what factors may lead to higher wisdom among the countries. Of course I do not want to investigate my time to Nobel winning social side, but more material side: IQ. Is it population, money or to my surprise the difference between heights of men and women?

To do that we might need 2 other datasets for knowing heights and IQ, and fastai and XGBoost for creating a model.
 - https://www.kaggle.com/datasets/majyhain/height-of-male-and-female-by-country-2022
 - https://www.kaggle.com/datasets/weavrk/average-iq-by-country

```python
heights=pd.read_csv('/kaggle/input/height-of-male-and-female-by-country-2022/Height of Male and Female by Country 2022.csv')
heights.rename(columns={'Country Name': 'Country'}, inplace=True)
heights.drop(['Rank', 'Male Height in Ft', 'Female Height in Ft'],
             axis=1, inplace=True)
```

```python
iq=pd.read_csv('/kaggle/input/average-iq-by-country/csvData.csv')
iq.rename(columns={'country': 'Country'}, inplace=True)
iq.drop(['rank', 'pop2022'], axis=1, inplace=True)
```

```python
df3=pd.merge(heights, iq, on='Country', how='left')
df3['difference_in_heights']=abs(df3['Male Height in Cm']-df3['Female Height in Cm'])
```

```python
df = pd.merge(df, df3, on='Country', how='left')
df.drop(['Male Height in Cm',	'Female Height in Cm'], axis=1, inplace=True)
df.set_index('Country', inplace=True)
```

```python
df.columns
```

```python
from fastai.tabular.all import *
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
```

```python
modes=df.mode().iloc[0]
df.fillna(modes, inplace=True)
```

```python
dep_var='iq'
cont, cat=cont_cat_split(df, 1, dep_var=dep_var)
```

```python
cat=['Continent']
cont=['2022 Population', 'Area','Destiny','Growth Rate', 'happiness','GDP per capita',
      'Social support', 'Healthy life expectancy','Freedom to make life choices','Generosity',
    'Perceptions of corruption','difference_in_heights']
```

```python
df3=df[cat+cont+[dep_var]]
```

```python
to=TabularPandas(df3, procs=[Categorify, Normalize], 
                cat_names=cat,cont_names=cont, y_names=dep_var, 
                splits=RandomSplitter(seed=25)(range_of(df3)), 
                 y_block=RegressionBlock())    
train_xs, train_y=to.train.xs, to.train.y
valid_xs, valid_y=to.valid.xs, to.valid.y
```

```python
model = XGBRegressor()
model.fit(train_xs, train_y);
mean_squared_error(model.predict(valid_xs), valid_y, squared=False)
```

Thank it pretty good model. Let's see feature importances

```python
from xgboost import plot_importance
plot_importance(model);
```

Average difference in heights of a men and women is more important that GDP, happiness and social support. XGBoost you are drunk, go home!
